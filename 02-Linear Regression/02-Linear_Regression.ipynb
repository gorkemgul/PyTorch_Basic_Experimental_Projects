```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch 
from torch.autograd import Variable
import torch.nn as nn
import warnings 
warnings.filterwarnings("ignore")
```

<a id = "2"></a>
## Data Preparation

```python
#Car Prices Data Preparation
car_prices_array = [3, 4, 5, 6, 7, 8, 9]
car_prices_np = np.array(car_prices_array, dtype = np.float32)
car_prices_np = car_prices_np.reshape(-1, 1)
car_prices_tensor = Variable(torch.from_numpy(car_prices_np))

#Number of Car Sells Data Preparation
num_of_car_sell_array = [7.5, 7, 6.5, 6.0, 5.5, 5.0, 4.5]
num_of_car_sell_np = np.array(num_of_car_sell_array, dtype = np.float32)
num_of_car_sell_np = num_of_car_sell_np.reshape(-1, 1)
num_of_car_sell_tensor = Variable(torch.from_numpy(num_of_car_sell_np))

#Visualization of Data 
plt.scatter(car_prices_array, num_of_car_sell_array, color = "green")
plt.title("Car Prices vs Number of Sells")
plt.xlabel("Car Prices")
plt.ylabel("Number of Car Sells")
plt.show()
```

<a id = "3"></a>
## Creating Model

```python
#Creating Class
class LinearRegModel(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegModel, self).__init__()
        
        self.linear = nn.Linear(input_dim, output_dim)
    
    def forward(self, x):
        return self.linear(x)
    
#Creating Model
input_dim = 1
output_dim = 1
model = LinearRegModel(input_dim, output_dim)

#MSE 
error = nn.MSELoss()

#Optimization
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(), learning_rate)
```

<a id = "4"></a>
## Model Training

```python
#Model Training
loss_list = []
n_iters = 1460

for iteration in range(n_iters):
    optimizer.zero_grad()
    results = model(car_prices_tensor)
    loss = error(results, num_of_car_sell_tensor)
    loss.backward()
    optimizer.step()
    loss_list.append(loss.data)
    
    if iteration % 50 == 0:
        print("iteration: {}, loss: {}".format(iteration, loss.data))

#Visualization of Loss
plt.plot(range(n_iters), loss_list)
plt.title("Loss Graphic")
plt.xlabel("Number of Iteration")
plt.ylabel("Loss")
plt.show()
```

<a id = "5"></a>
## Prediction and Visualization

```python
predicted = model(car_prices_tensor).data.numpy()
plt.scatter(car_prices_array, num_of_car_sell_array, label = "Data", color = "green")
plt.scatter(car_prices_array, predicted, label = "Predicted Data", color = "purple")
plt.title("Predicted Data vs Original Data")
plt.xlabel("Car Prices")
plt.ylabel("Number of Car Sells")
plt.show()
```
